```python
# drop unnecessary co
drop_columns = ufo_df.drop(['comments', 'duration (hours/min)', 'duration (seconds)', 'date posted' ], axis=1)
```

```python
test = drop_columns.copy()
```

```python
us_only = test.dropna(subset=['state'])
```

```python
us_only
```

```python
us_only['state'].value_counts()
```

```python
# create a dataframe with only California data
cali_only_df = us_only.loc[us_only['state'] == 'ca']
```

```python

cali_only_df
```

```python
#duplicate datetime column

cali_only_df['datetime_copy'] = cali_only_df['datetime']
```

```python
cali_only_df
```

```python
# split the datetime copy column to get a date column
cali_only_df['date'] = cali_only_df['datetime_copy'].str.split(' ').str[0]
```

```python
cali_only_df
```

```python
# split datetime copy column to get a year column
cali_only_df['year'] = cali_only_df['datetime_copy'].str.split('-').str[0]
```

```python
#keep only data from 2003-2013
cali_decade = cali_only_df.copy()

years = ['2013','2012','2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004', '2003']

cali_decade = cali_decade[cali_decade["year"].isin(years)==True]
```

```python
cali_decade
```

```python
# convert datetime to datetime datatype

cali_decade["date"] = cali_decade['date'].astype('datetime64[ns]')
```

```python
# drop extra datetime column
cali_decade = cali_decade.drop(['datetime_copy'], axis=1)
```
